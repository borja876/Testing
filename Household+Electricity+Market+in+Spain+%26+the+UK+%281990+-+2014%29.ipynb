```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
w = pd.read_csv('D:/Users/Borja.gonzalez/Desktop/Bootcamp Prep/UNdata_Export_20170628_003608818.csv')
```

```python
df = pd.DataFrame(w)
```

```python
df
```

```python
df1 = df.ix[lambda df: w['Country or Area'] == "Spain", :]
df1
```

```python
plt.scatter(x = df1["Year"], y = df1['Quantity'], color = 'purple', marker = 'x', s = 10)
plt.ylabel('Kilowatt-hours, million')
plt.ylim([20000, 160000])
plt.title('Electricity consumption in Spain by household 1990-2014')
plt.show()
```

There is a sustained growth in the electricity consumption in Spain from 1990 to 2014. This is a good indicator of the economic growth of the country although between 2005 and 2015 there is a decrease in the interannual grouwth due to aggressive energy efficiency measures.

```python
df2 = df.ix[lambda df: w['Country or Area'] == "United Kingdom", :]
df2
```

```python
plt.scatter(x = df2["Year"], y = df2['Quantity'], color = 'green', marker = 'x', s = 10)
plt.ylabel('Kilowatt-hours, million')
plt.ylim([20000, 160000])
plt.title('Electricity consumption in UK by household 1990-2014')
plt.show()
```

The electricity consumption experiments a moderate growth from 1990 to 2015. There is a higher growth between 1990 and 2005 than from 2005 onwards. This is due to the introduction of economic incentives to reduce household electricity consumption at a European level.

```python
plt.figure(figsize=(20, 5))

plt.subplot(1, 3, 1)
plt.scatter(x = df1["Year"], y = df1['Quantity'], color = 'purple', marker = 'x', s = 10)
plt.ylabel('Kilowatt-hours, million')
plt.ylim([20000, 160000])
plt.title('Electricity consumption in Spain by household 1990-2014')

plt.subplot(1, 3, 2)
plt.scatter(x = df2["Year"], y = df2['Quantity'], color = 'green', marker = 'x', s = 10)
plt.ylabel('Kilowatt-hours, million')
plt.ylim([20000, 160000])
plt.title('Electricity consumption in UK by household 1990-2014')

plt.subplot(1, 3, 3)
plt.scatter(x = df1["Year"], y = df1['Quantity'], color = 'purple', marker= "x", s=10)
plt.scatter(x = df2["Year"], y = df2['Quantity'], color = 'green', marker ="x", s=10)
plt.ylabel('Kilowatt-hours, million')
plt.ylim([20000, 160000])
plt.title('Electricity consumption in Spain, UK by household 1990-2014')

plt.tight_layout()
plt.show()
```

The electricity consumption in the UK is higher than Spain due to weather conditions. Although there is a growth in the electricity consumption, in relative terms is lower than in Spain and there is a decline from 2005 to 2015 due to the economic incentives enforced in Europe for the electricity sector. Although Spain continues to grow during that period, the UK starts a decline due to lower natural gas prices.
