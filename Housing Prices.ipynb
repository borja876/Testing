```python
#Create the clusters of neighbouhoods/suburbs as it will give the adequate level of granularity
#Consider prices and frequency of purchase in each area
freq = housing4.groupby('Suburb').count()['Postcode']
mean = housing4.groupby('Suburb').median()['Price_per_buildingarea']
cluster = pd.concat([freq, mean], axis=1)
cluster['Suburb'] = cluster.index
cluster.columns = ['freq', 'Price_per_buildingarea','Suburb']
cluster.describe()
```

```python
#Divide into 2 clusters: Moderate and Expensive according to price. Create the moderate cluster
cluster1 = cluster[cluster.Price_per_buildingarea <= 6131]
cluster1.index
```

```python
#Create the Expensive cluster
cluster2 = cluster[cluster.Price_per_buildingarea > 6131]
cluster2.index
```

```python
#Add the clusters into the dataset
def get_group(x):
    if x in cluster1.index:
        return 'Moderate'
    else:
        return 'Expensive'
housing4['Neighbourhood'] = housing4.Suburb.apply(get_group)
```

***Feature Selection using Lasso***

```python
#Feature Selection using Lasso

plt.figure(figsize=(20, 5))
sns.set_style("whitegrid")

plt.subplot(1, 4, 1)
for lambd in [x * 0.01 for x in range(1, 100)]:
    lasso = Lasso(alpha=lambd)
    lasso_coef = lasso.fit(X0, Y).coef_
    plt.xticks(range(len(names)), names, rotation=90)
    plt.ylabel('Coefficients')
    plt.plot(range(len(names)), lasso_coef)
    plt.title('Lasso (All features)')

plt.subplot(1, 4, 2)
for lambd in [x * 0.01 for x in range(1, 100)]:
    lasso = Lasso(alpha=lambd)
    lasso_coef = lasso.fit(XRFE, Y).coef_
    plt.xticks(range(len(namesRFE)), namesRFE, rotation=90)
    plt.ylabel('Coefficients')
    plt.plot(range(len(namesRFE)), lasso_coef)
    plt.title('Lasso RFE')

plt.subplot(1, 4, 3)    
for lambd in [x * 0.01 for x in range(1, 100)]:
    lasso = Lasso(alpha=lambd)
    lasso_coef = lasso.fit(XBS, Y).coef_
    plt.xticks(range(len(namesBS)), namesBS, rotation=90)
    plt.ylabel('Coefficients')
    plt.plot(range(len(namesBS)), lasso_coef)
    plt.title('Lasso (Best Selector)')
    
plt.subplot(1, 4, 4)
for lambd in [x * 0.01 for x in range(1, 100)]:
    lasso = Lasso(alpha=lambd)
    lasso_coef = lasso.fit(XFIRF, Y).coef_
    plt.xticks(range(len(namesFIRF)), namesFIRF, rotation=90)
    plt.ylabel('Coefficients')
    plt.plot(range(len(namesFIRF)), lasso_coef)
    plt.title('Lasso on Random Forest')
    
plt.tight_layout()
plt.show()
```

```python
#New set of features based on results obtained in with Lasso, Random Forest and RFE

Xl = X1[['Rooms', 'BuildingArea', 'Regionname','Distance', 'Neighbourhood']]
```

```python
#Split the data into training and testing datasets. Split: 70/30; train/test

X_train, X_test, y_train, y_test = train_test_split(Xl,Y, test_size=0.3, random_state=0)

#Initiating the cross validation generator, N splits = 5
kf = KFold(5)
```

***Linear Regression***

```python
#Fit the Linear Regression model with the PCA features
lr = LinearRegression()

lr.fit(X_train, y_train)

print(lr.coef_)
```

```python
#Predict the value of y
y_pred = lr.predict(X_test)

#Mean squared error of the regressor
lin_mse = mean_squared_error(y_pred, y_test)
lin_rmse = np.sqrt(lin_mse)

#Accuracy of the regressor
accln = cross_val_score(lr, X_test, y_test, cv=kf).mean()
acclnpca = cross_val_score(lr, XPCA, Y, cv=kf).mean()

print('RMSE: %.4f' % lin_rmse)
print('Linear Regression Accuracy: %.4f' % accln)
print('Linear Regression Accuracy PCA: %.4f' % acclnpca)
```

***Ridge Regression***

```python
# Testing the Ridge Regression.

ridgeregr = Ridge()

k1 = np.arange(50)+1

parameters = {'alpha': k1}

#Fit parameters

ridgeregr1 = GridSearchCV(ridgeregr, param_grid=parameters, cv=kf)


#Fit the tunned classifier in the traiing space
ridgeregr1.fit(X_train, y_train)

#Print the best parameters
print(ridgeregr1.best_params_)
```

```python
# Fit the model.
ridgeregr1.fit(X_test, y_test)

#Predict the value of y
y_pred = ridgeregr1.predict(X_test)

#Mean squared error of the regressor
rid_mse = mean_squared_error(y_pred, y_test)
rid_rmse = np.sqrt(rid_mse)

#Accuracy of the regressor
accridg = cross_val_score(ridgeregr1, X_test, y_test, cv=kf).mean()
accridgpca = cross_val_score(ridgeregr1, XPCA, Y, cv=kf).mean()

print('RMSE: %.4f' % rid_rmse)
print('Ridge Regression Accuracy: %.4f' % accridg)
print('Ridge Regression Accuracy PCA: %.4f' % accridgpca)
```

***Random Forest***

```python
#Initialize and fit the model

abc = RandomForestRegressor(n_jobs=-1)

#Create range of values to fit parameters

k2 = ['mse','mae']
k4 = ['auto','log2']
k3 = [150, 200,250, 300]

parameters = { 'criterion': k2,
              'n_estimators': k3,
             'max_features': k4
}

#Fit parameters

clf = GridSearchCV(abc, param_grid=parameters, cv=kf)

clf.fit(X_train,y_train)

#The best hyper parameters set
print("Best Hyper Parameters:", clf.best_params_)
```

The Melbourne real estate market 2016-2017 prices have been analyzed in the dataset. The goal is to create a predictive model that gives the most important features that have influence the final price of real estate. As a real estate agent that will help to focus, the efforts on those assets that have the most meaningful features before the go into the market and while they are in the market. This also will help the agent to accurately know the price of the asset when it is bought or resold to the final customer and indirectly predict the margins.

To reduce the number of outstanding values in the dataset, the categorical values for the “Regionname” and “CouncilArea” have been dropped. For the rest of them, the median of the existing values have been used.

The price has been set up as the output feature being the rest predictors. During the data cleansing, the price has been transformed into the “logprice” to smooth the skewness it presented towards low prices.
The feature price_per_buidlingarea has been created. Although it was expected to have a higher correlation with the price, it doesn´t and it has been used to cluster neighbourhoods into expensive and moderate according to the price of assets in those neighbourhoods.

The initial features: suburb, council area longtitude and latitude have been merged into a Neighbourhood feature in which two neighbouhoods have been considered according to the price_per_building area.
The yearbuilt has been transformed into “Age” considering only those assets built after 1970 to eliminate the outliers found during the initial data exploration. Additionally, those houses with more than 16 rooms have been excluded as outliers that were penalizing the accuracy of the model.

From a PCA analysis there are only six components that explain most part of the variance (up to 85%). Hence to select the features the RFE, Best Select Random Forest Feature Importance and Lasso regression have been carried out considering only six features. All of them give the same features in different orded, except for the price_per_buildingarea that appears as an important feature only in the random forest feature importance methodology.

The chosen features are: 
Xl = X1[['Rooms', 'BuildingArea', 'Regionname','Distance', 'Neighbourhood']]

These features have been used in the Linear Regression, Ridge Regression, Random Forest, Decision Tree, Adam Boost, Gradient Boosting Regressor, KNN Regressor and SVR. In the training set the models´ hyper parameters have been tuned using gridsearch and the final values run on the test set have been compared against the PCA result. 

The accuracy obtained in the test sets after the models have been tuned are:
Linear Regressor: 0.6702
Ridge Regressor: 0.6699
Random Forest: 0.8021
Decision Tree: 0.6456
Adam Boost Regressor: 0.7215
Gradient Boosting Regressor: 0.7972
KNN Regressor: 0.7738
SVR: 0.7805

In all of the above mentioned models  the Root mean square error has been evaluated following the same pattern than the accuracy with a min value of 0.4282 when logprices (predicted and real) are compared.
